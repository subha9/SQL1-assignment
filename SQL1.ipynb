```python
import numpy as np
import pandas as pd
from pandasql import sqldf
```

```python
sqladb = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
```

```python

sqladb.rename(index=str, columns={'39':'age',' State-gov':'workclass',' 77516':'fnlwgt',' Bachelors':'education',' 13':'education-num',' Never-married':'marital-status',' Adm-clerical':'occupation',' Not-in-family':'relationship',' White':'race',' Male':'sex',' 2174':'capital-gain',' 0':'capital-loss',' 40':'hours-per-week',' United-States':'native-country'}, inplace=True)
```

```python
sqladb
```

```python
sqldf("SELECT * FROM sqladb LIMIT 0,10")
```

```python
#2. Show me the average hours per week of all men who are working in private sector
sqldf("SELECT AVG([hours-per-week]) AS Hours FROM sqladb WHERE TRIM(workclass)='Private' AND TRIM(sex) = 'Male'")
```

```python
sqldf("Select COUNT (education) AS [Education-frequency],COUNT(occupation) AS [occupation-frequency],COUNT(relationship)AS[relationship-frequency]from sqladb")
```

```python
sqldf("SELECT COUNT(*) AS [married-private-masters] FROM sqladb WHERE ([marital-status] NOT LIKE 'Never-married' OR [marital-status] NOT LIKE 'divorced') AND workclass = 'Private' AND education LIKE 'Masters'")
```

```python
sqldf("SELECT AVG(age) AS Average, MIN(age) AS Minimum, MAX(age) AS Maximum FROM sqladb GROUP BY workclass")
```

```python
sqldf("SELECT SUM(age) AS [age-distribution], [native-country] FROM sqladb GROUP BY [native-country]")
```

```python
sqldf("SELECT ABS([capital-gain] - [capital-loss]) AS [Net-capital-gain] FROM sqladb")
```
